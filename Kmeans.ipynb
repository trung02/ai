```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

features, true_labels = make_blobs(n_samples=400,n_features=2,centers=4,cluster_std=2.75,random_state=42)
#plt.scatter(features[:,0],features[:,1])

kmeans = KMeans(n_clusters=4, random_state=0).fit(features)
color_dict = {0:'orange',1:'g',2:'b',3:'y'}
colors = [color_dict[i] for i in kmeans.labels_]
plt.scatter(features[:,0],features[:,1], c = colors)
centroids = [kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1]]
print('centroids: ',centroids)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], c = 'r')
plt.show()
```
